```python
# train.py
from ultralytics import YOLO

def main():
    # Load YOLOv8 nano model
    model = YOLO("yolov8n.pt")   # pretrained on COCO, good starting point

    # Train on PPE dataset
    model.train(
        data="data.yaml",     # path to your data.yaml
        epochs=50,            # adjust higher for better results (100â€“300 typical)
        imgsz=640,            # training image size
        batch=-1,             # auto batch size based on GPU
        project="ppe_runs",   # results folder
        name="yolov8n_baseline", # experiment name
        device=0              # use GPU if available
    )

    # Evaluate on test set
    metrics = model.val(split="test")
    print(metrics)

if __name__ == "__main__":
    main()
```

```python
#Agumenting the gloves sample#
from ultralytics import YOLO

def main():
    model = YOLO("yolov8n.pt")

    model.train(
        data="data.yaml",
        epochs=100,           # train longer
        imgsz=640,
        batch=-1,
        project="ppe_runs",
        name="yolov8n_aug_gloves",
        device=0,
        # Augmentations
        hsv_h=0.015,          # Hue augmentation
        hsv_s=0.7,            # Saturation
        hsv_v=0.4,            # Brightness
        degrees=10,           # Rotation
        translate=0.1,        # Translation
        scale=0.5,            # Zoom in/out
        shear=2.0,            # Shear
        perspective=0.0005,   # Slight perspective
        flipud=0.0,           # No vertical flip
        fliplr=0.5,           # Random horizontal flip
        mosaic=1.0,           # Strong mosaic
        mixup=0.2,            # Mixup augmentation
        copy_paste=0.2        # Copy-paste objects
    )

if __name__ == "__main__":
    main()
```
